```python
##### separate x and y, train and test

#### data fb

x_all = fb.copy()
# print(x_all)

x = x_all.drop(columns=['User_ID','Product_ID','Purchase'])
y = x_all['Purchase']
# print(x)
# print(y)

##### convert to numpy
x_np = x.to_numpy().astype('float64')
y_np = y.to_numpy().astype('float64')
print(x_np)
print(y_np)

y_np.resize((len(y_np),1))
print(y_np)

# x_np_norm = normalize(x_np)
# y_np_norm = normalize(y_np)
# print(x_np_norm)
# print(y_np_norm)

scaler = MinMaxScaler()

x_np_scaled = scaler.fit_transform(x_np)
y_np_scaled = scaler.fit_transform(y_np)

print(x_np_scaled)
print(y_np_scaled)

##### split into x_tr, x_te, y_tr, y_te
x_train, x_test, y_train, y_test = train_test_split(x_np_scaled,y_np_scaled,train_size=0.90)
print(x_train.shape,y_train.shape)
print(x_test.shape,y_test.shape)
print(y_test)
```

```python
model = keras.Sequential(
    [
        layers.Dense(128, activation="relu", input_shape=(x_train.shape[-1],)),
        layers.Dense(256, activation="relu"),
        layers.Dense(128, activation="relu"),
        layers.Dense(64, activation="relu"),
        layers.Dense(1, activation="relu"),
    ]
)
model.summary()

model.compile(optimizer=keras.optimizers.SGD(learning_rate=0.05, momentum=0.04, nesterov=False, name="SGD"), 
              loss=keras.losses.MeanSquaredError(),
             metrics=[keras.metrics.SparseCategoricalAccuracy(name="acc")])

model.fit(x_train,y_train,batch_size=41255,epochs=100)

print('val')
model.fit(x_test,y_test,batch_size=440,epochs=1)

model.layers[2].output.get_shape()
```
